```python
from ipywidgets import interact, widgets
import numpy as np
import matplotlib.pyplot as plt

def simula_arrhenius(Ea):
  k0 = 7.75e9
  R = 8.314
  T = np.linspace(273.15, 353.15)
  k = k0*np.exp(-Ea/(R*T))
  plt.plot(T,k,'o-b', label=f'Ea = {Ea:1.4f}')
  plt.grid()
  plt.xlabel('Temperatura, T')
  plt.ylabel('Constante cinética, k')
  plt.legend()
  plt.show()

interact(simula_arrhenius,
         Ea = widgets.FloatSlider(min=40e3, max=60e3, step=100, value=50e3));
```
